```python
import numpy as np
import pandas as pd

file_path1 = '/content/dump_report_2024-04-21.xlsx'

df = pd.read_excel(file_path1)

df.head()
```

```python
print(df.shape)
complaint_data = df[df['TicketType'] == 'Complaint']
print(complaint_data.shape)
complaint_data1= complaint_data
```

```python
complaint_data.drop(["comment_template", "ResolutionName","WorkflowName","FromStatusName","ToStatusName","Comments","BuildingInstallationShopTypeFo","CallingNumberDetails","ChargedAmount","CustomerEmail","ErrorPrompt","EscalationAccuracy","FeedbackForTicketCreator","MonthlyYearlyRevenueInformatio","PrepaidLifecycleStatus","Prompt","SIMNumber","SMSFeedbackToCustomer","SMSToAlternateNumber","ThreeMonthsAverageUsageForEach"], axis=1, inplace=True)
```

```python
print(complaint_data.head(1))
```

```python
from datetime import datetime
complaint_data1['TaskAssignDate'] = pd.to_datetime(complaint_data1['TaskAssignDate'])
complaint_data1['TaskCompletionDate'] = pd.to_datetime(complaint_data1['TaskCompletionDate'], errors='coerce')

complaint_data1['TaskDuration'] = datetime.now() - complaint_data1['TaskAssignDate']
complaint_data1.loc[complaint_data1['TaskCompletionDate'].notnull(), 'TaskDuration'] = (complaint_data1['TaskCompletionDate'] - complaint_data['TaskAssignDate'])


complaint_data1['TaskDuration'] = complaint_data1['TaskDuration'].dt.total_seconds() / (60 * 60 * 24)

complaint_data1['TotalDuration'] = complaint_data1.groupby('TicketId')['TaskDuration'].transform('sum')

complaint_data1['TotalDuration'] = complaint_data1['TotalDuration'].round().astype('Int64')
complaint_data['TaskDuration'] = complaint_data1['TaskDuration']
complaint_data['TotalDuration'] = complaint_data1['TotalDuration']
```

```python
columns_to_modify = ["comment_template", "ResolutionName", "WorkflowName", "FromStatusName", "ToStatusName", "Comments", "BuildingInstallationShopTypeFo", "CallingNumberDetails", "ChargedAmount", "CustomerEmail", "ErrorPrompt", "EscalationAccuracy", "FeedbackForTicketCreator", "MonthlyYearlyRevenueInformatio", "PrepaidLifecycleStatus", "Prompt", "SIMNumber", "SMSFeedbackToCustomer", "SMSToAlternateNumber", "ThreeMonthsAverageUsageForEach"]
for column in columns_to_modify:
    complaint_data[column] = ''
```

```python
complaint_data['TechOrNonTech'] = complaint_data.groupby('TicketId')['tag2'].transform(lambda x: 'Tech' if 'Technology' in x.values else 'Non Tech')
complaint_data2= complaint_data
```

```python
#print(complaint_data2.head())
```

```python
from datetime import datetime
complaint_data2['TotalMTTR'] =0

complaint_data2['TaskAssignDate'] = pd.to_datetime(complaint_data2['TaskAssignDate'])
complaint_data2['TaskCompletionDate'] = pd.to_datetime(complaint_data2['TaskCompletionDate'], errors='coerce')
complaint_data2['TicketClosedDate'] = pd.to_datetime(complaint_data2['TicketClosedDate'], errors='coerce')

complaint_data2['TaskDuration1'] = complaint_data2['TicketClosedDate'] - complaint_data2['TaskAssignDate']
complaint_data2.loc[complaint_data2['TaskCompletionDate'].notnull(), 'TaskDuration1'] = (complaint_data2['TaskCompletionDate'] - complaint_data2['TaskAssignDate'])


complaint_data2['TaskDuration1'] = complaint_data2['TaskDuration1'].dt.total_seconds() / (60 * 60 * 24)

complaint_data2['TotalMTTR'] = complaint_data2.groupby('TicketId')['TaskDuration1'].transform('sum')

complaint_data2['TotalMTTR'] = complaint_data2['TotalMTTR'].round().astype('Int64')
complaint_data['TaskDuration1'] = complaint_data1['TaskDuration1']
complaint_data['TotalMTTR'] = complaint_data2['TotalMTTR']
```
